# Árboles de Regresión 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Position_Salaries.csv")
dataset
```

## Pre Procesado de Datos

```python
X = dataset.iloc[:, 1:2].values
y = dataset.iloc[:, 2].values
```

## Escalado de Datos

En este caso no es necesario


## Ajustar la regresión con los datos

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
regression = DecisionTreeRegressor(random_state = 0)
regression.fit(X, y)
```

### Haciendo predicciones

```python
y_pred = regression.predict([[6.5]])
y_pred
```

```python
plt.figure(figsize = (15,15))
plt.scatter(X,y, color = "red")
plt.plot(X, regression.predict(X), color = "blue")
plt.title("Modelo de Regresión (Árbol de Regresión) con valores escalados")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.figure(figsize = (15,15))
plt.scatter(X,y, color = "red")
plt.axhline(y = regression.predict([[6.5]]), color = "brown")
plt.axvline(x = 6.5, color = "brown")
plt.plot(X_grid, regression.predict(X_grid), color = "blue")
plt.title("Modelo de Regresión (SVR) con valores escalados")
plt.xlabel("Posición del empleado")
plt.ylabel("Sueldo (en $)")
plt.show()
```

Se crea un nodo que tiene un único elemento, de tal forma que al querer predecir, solo toma el más cercano por ello. Por eso al nivel de empleado 6.5 se le asigna el valor del nivel 6. Este algoritmo se mejora con la regresión de bosques aleatorios.
